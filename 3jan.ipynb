```python
area = float(input("Enter area in acres: "))
yield_per_acre = float(input("Enter yield per acre (in kg): "))
total_yield = area * yield_per_acre
print("Total crop yield is:", total_yield, "kg")
```

```python
field1 = float(input("Enter acres of field 1: "))
field2 = float(input("Enter acres of field 2: "))
difference = abs(field1 - field2)
print("Difference in acres:", difference)
```

```python
cost = float(input("Enter cost of production: "))
revenue = float(input("Enter selling price: "))
if revenue > cost:
    print("Profit:", revenue - cost)
elif revenue < cost:
    print("Loss:", cost - revenue)
else:
    print("No Profit, No Loss")
```

```python
stages = ["Germination", "Vegetative", "Flowering", "Maturity", "Harvest"]
for stage in stages:
    print("Crop is in", stage, "stage")
```

```python
days = int(input("Enter total days: "))
interval = int(input("Enter irrigation interval (days): "))
print("Irrigation Days:")
for day in range(1, days + 1):
    if day % interval == 0:
      print("Irrigate on Day", day)
```

```python
area = float(input("Enter area in acres: "))
rate = float(input("Enter fertilizer rate per acre (kg): "))
total_fertilizer = area * rate
print("Total fertilizer needed:", total_fertilizer, "kg")
```

```python
last_service_hours = int(input("Enter last service hour reading: "))
current_hours = int(input("Enter current hour reading: "))
if current_hours - last_service_hours >= 250:
    print("Service Due!")
else:
    print("Service not yet due.")
```
